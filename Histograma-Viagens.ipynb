```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(20,10)})
sns.set_theme(style="whitegrid")
```

```python
DATASETS_DIR = '../dados/subprefeituras/'
DATASETS = ['São Mateus', 'Vila Mariana', 'Jabaquara', 'Cidade Ademar', 'Itaim Paulista', "M'Boi Mirim", 'Pinheiros', 'São Miguel', 'Capela do Socorro', 'Vila Maria/Vila Guilherme', 'Sé', 'Santo Amaro', 'Lapa', 'Freguesia do Ó', 'Campo Limpo', 'Guaianases', 'Penha', 'Cidade Tiradentes', 'Aricanduva', 'Jaçanã', 'Perus', 'Sapopemba', 'Casa Verde', 'Mooca', 'Vila Prudente', 'Ipiranga', 'Parelheiros', 'Ermelino Matarazzo', 'Santana', 'Butantã', 'Pirituba', 'Itaquera']

dfs = {df_name: pd.read_csv(f"{DATASETS_DIR}{df_name.replace(' ', '_').replace('/', '_')}.csv") for df_name in DATASETS}

all_data = pd.concat(dfs.values(), axis=0, ignore_index=True)
male = all_data.loc[all_data['SEXO'] == 1.0]
female = all_data.loc[all_data['SEXO'] == 2.0]
```

# Trips

```python
plt.figure(figsize=[20,10], facecolor='white')
g = sns.histplot(data=all_data, weights=all_data['FE_VIA'].values, x="H_SAIDA", color='gray', kde=True, discrete=True, stat='percent')
g.set(xlim=(0, 23), 
    ylim=(0, 35),
    xlabel="Hora de Saída",
    ylabel="Porcentagem de Viagens",
    xticks=range(24),
    title=f"Porcentagem de Viagens por H_SAIDA")
plt.savefig(f"./output/1.png")
```

```python
plt.figure(figsize=[20,10], facecolor='white')
g = sns.histplot(data=male, weights=male['FE_VIA'].values, x="H_SAIDA", kde=True, discrete=True, stat='percent')
g.set(xlim=(0, 23), 
    ylim=(0, 35),
    xlabel="Hora de Saída",
    ylabel="Porcentagem de Viagens",
    xticks=range(24),
    title=f"Porcentagem de Viagens por H_SAIDA (SEXO=MASCULINO)")
plt.savefig(f"./output/2.png")
```

```python
plt.figure(figsize=[20,10], facecolor='white')
g = sns.histplot(data=female, weights=female['FE_VIA'].values, x="H_SAIDA", color='red', kde=True, discrete=True, stat='percent')
g.set(xlim=(0, 23), 
    ylim=(0, 35),
    xlabel="Hora de Saída",
    ylabel="Porcentagem de Viagens",
    xticks=range(24),
    title=f"Porcentagem de Viagens por H_SAIDA (SEXO=FEMININO)")
plt.savefig(f"./output/3.png")
```
